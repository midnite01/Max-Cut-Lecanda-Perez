##Definicion del problema

El problema Max-Cut consiste en particionar los vértices de un grafo en dos subconjuntos de forma que se maximice el peso de las aristas que cruzan entre ellos. Este problema pertenece a la clase NP-completo, lo que implica que, aunque una solución puede verificarse en tiempo polinomial, no se conoce ningún algoritmo eficiente que lo resuelva en todos los casos. Por ello, en la práctica se emplean tanto métodos exactos como heurísticos para abordar instancias de tamaño considerable.

###Descripción en lenguaje natural

El problema de Max cut analiza un grafo y elige la mejor forma de dividir los nodos en dos grupos tal que las aristas que cruzan entre ellos se maximicen.

### Definicin formal (entraga/salida)

ENTRADA:
Un grafo G = (V, E) donde “V” es el conjunto de vértices y “E” el conjunto de aristas que conectan pares de vértices. Opcionalmente, cada arista puede tener un peso asociado.

SALIDA:
Una partición del conjunto de vértices “V” en dos subconjuntos disjuntos, tal que se maximice la cantidad (o el peso) de las aristas que conectan vértices pertenecientes a distintos subconjuntos.

##Aplicaciones

El problema Max-Cut tiene diversas aplicaciones en áreas donde se requiere analizarla estructura de redes o sistemas interconectados, algunos ejemplos de esto son.

1.-Redes sociales: En redes sociales, los nodos pueden representar usuarios y las aristas interacciones entre ellos. Max-Cut permite identificar particiones que maximizan las interacciones entre grupos, lo que puede evidenciar fenómenos como polarización o conflicto entre comunidades.


2.- Diseño de circuitos (VLSI): En el diseño de circuitos electrónicos, Max-Cut se utiliza para particionar componentes de manera que se analicen o gestionen las conexiones entre distintas partes del sistema.


3.- Computación paralela: En sistemas distribuidos, los nodos pueden representar tareas y las aristas dependencias o comunicaciones. Max-Cut permite estudiar cómo dividir el sistema en dos partes para analizar el costo de comunicación entre ellas.





##Modelamiento matemático

El problema Max Cut consiste en dividir los vértices de un grafo no dirigido en dos subconjuntos disjuntos, de manera que se maximice la cantidad (o peso) de aristas que conectan vértices pertenecientes a subconjuntos distintos.

Sea un grafo:

$$
G = (V, E)
$$

donde $ V $ es el conjunto de vértices y $ E $ es el conjunto de aristas.

###Conjuntos y parámetros

Se definen los siguientes elementos:

- $ V $: conjunto de vértices del grafo  
- $ E $: conjunto de aristas del grafo  

Cada arista tiene asociado un peso:

$$
w_{ij}
$$

Si el grafo es no ponderado, se considera:

$$
w_{ij} = 1 \quad \forall (i,j) \in E
$$

Este modelo permite trabajar tanto con grafos ponderados como no ponderados.


###Variables de decisión

Se define una variable binaria para cada vértice:

$$
x_i \in \{0,1\}, \quad \forall i \in V
$$


Interpretación:

- $ x_i = 0 $ → vértice en el conjunto A  
- $ x_i = 1 $ → vértice en el conjunto B  

Además, se define una variable binaria para cada arista:

$$
y_{ij} \in \{0,1\}, \quad \forall (i,j) \in E
$$

Donde:

- $ y_{ij} = 1 $: la arista $i,j$ pertenece al corte  
- $ y_{ij} = 0 $: la arista no pertenece al corte  

Una arista pertenece al corte cuando sus extremos están en subconjuntos distintos, es decir:

$$
x_i \neq x_j
$$


###Restricciones

Para modelar correctamente la relación entre $ x_i $, $ x_j $ y $ y_{ij} $, se imponen las siguientes restricciones:

$$
y_{ij} \leq x_i + x_j
$$

$$
y_{ij} \leq 2 - x_i - x_j
$$

$$
y_{ij} \geq x_i - x_j
$$

$$
y_{ij} \geq x_j - x_i
$$

Estas restricciones permiten representar la siguiente relación:

$$
y_{ij} = |x_i - x_j|
$$

De esta forma:

- Si $ x_i = x_j $, entonces $ y_{ij} = 0 $  
- Si $ x_i \neq x_j $, entonces $ y_{ij} = 1 $  

Finalmente, se establecen las restricciones de dominio:

$$
x_i \in \{0,1\}, \quad \forall i \in V
$$

$$
y_{ij} \in \{0,1\}, \quad \forall (i,j) \in E
$$


###Función objetivo

El objetivo es maximizar la suma de los pesos de las aristas que pertenecen al corte:

$$
\max Z = \sum_{(i,j)\in E} w_{ij} \cdot y_{ij}
$$

En el caso no ponderado, la función objetivo se reduce a:

$$
\max Z = \sum_{(i,j)\in E} y_{ij}
$$

Esto corresponde a maximizar la cantidad de aristas que cruzan entre los dos subconjuntos.


###Versión de decisión

La versión de decisión del problema Max Cut se formula como:

Dado un grafo $ G = (V,E) $ y un valor $ K $, determinar si existe una partición de los vértices en dos subconjuntos tal que:

$$
\sum_{(i,j)\in E} w_{ij} \cdot y_{ij} \geq K
$$

Si existe una partición que cumple esta condición, la respuesta es "sí". En caso contrario, la respuesta es "no".

En la versión no ponderada, esto equivale a determinar si existe un corte con al menos $ K $ aristas.

##Complejidad y contexto

###Justificacion de pertenencia a NP

Un problema pertenece a la clase NP si, dada una solución candidata, es posible verificar su validez en tiempo polinomial mediante una máquina determinista.

En el caso del problema Max-Cut, una solución consiste en una partición de los vértices del grafo en dos subconjuntos. Para verificar su calidad, basta con recorrer el conjunto de aristas y comprobar cuáles conectan vértices pertenecientes a distintos subconjuntos, acumulando el peso correspondiente.

Este procedimiento puede realizarse en tiempo $O(∣E∣)$, lo que es polinomial respecto al tamaño del grafo.

Por lo tanto, el problema Max-Cut pertenece a la clase NP.

###Explicacion conceptual de NP-completitud

La clase NP puede entenderse en términos de máquinas no deterministas, las cuales, de manera conceptual, son capaces de “adivinar” una solución y luego verificarla en tiempo polinomial mediante una máquina determinista.

En el caso de Max-Cut, una máquina no determinista podría generar una partición de los vértices y posteriormente verificar en tiempo polinomial el valor del corte asociado.

Un problema es NP-completo si cumple dos condiciones: pertenece a NP y es NP-hard, es decir, es al menos tan difícil como cualquier problema en NP. En este sentido, Max-Cut es NP-completo, ya que no se conoce un algoritmo determinista que lo resuelva en tiempo polinomial en todos los casos, y además otros problemas NP-completos pueden reducirse a él en tiempo polinomial.

Esto implica que encontrar una solución óptima requiere, en el peor caso, explorar un número exponencial de posibles particiones, lo que refleja la complejidad intrínseca del problema.

###Relacion con otros problemas

El problema Max-Cut se encuentra estrechamente relacionado con otros problemas clásicos de la teoría de la complejidad.

En particular, puede formularse como un problema de programación entera binaria, lo que lo vincula con el problema de 0–1 Integer Programming, también NP-completo.

Asimismo, presenta una relación conceptual con el problema Partition, en el cual se busca dividir un conjunto en dos subconjuntos equilibrados, aunque en Max-Cut el objetivo es maximizar las conexiones entre ellos.

Por otra parte, existe una diferencia relevante con el problema Min-Cut, el cual consiste en minimizar el número de aristas entre dos subconjuntos y puede resolverse en tiempo polinomial. Esta diferencia resalta cómo pequeños cambios en la formulación pueden implicar grandes diferencias en complejidad computacional.

Finalmente, Max-Cut también se relaciona con modelos de la física estadística, como el modelo de Ising, donde el problema puede interpretarse como la maximización de la energía de un sistema.

##Implementacion

###Decisión solucion base

La solución base implementada corresponde a un algoritmo de fuerza bruta. Este enfoque evalúa todas las particiones posibles del conjunto de vértices y calcula el valor del corte para cada una de ellas. La mejor partición encontrada corresponde a la solución óptima del problema para esa instancia.

En Max-Cut, una partición divide los vértices del grafo en dos subconjuntos. Sin embargo, una partición y su complemento representan el mismo corte. Por ejemplo, colocar los vértices $\{1,2\}$ en el conjunto $A$ y $\{3,4\}$ en el conjunto $B$ produce el mismo corte que invertir ambos grupos. Por esta razón, se fija un vértice en uno de los subconjuntos y solo se evalúan $2^{n-1}$ particiones en lugar de $2^n$.

La complejidad aproximada de este enfoque es:

$$
O(2^{n-1} \cdot m)
$$

donde $n$ representa el número de vértices y $m$ el número de aristas. Esta expresión se debe a que se evalúan $2^{n-1}$ particiones y, para cada una, se recorren las aristas del grafo para calcular cuántas cruzan entre ambos subconjuntos.

Aunque este algoritmo no es eficiente para instancias grandes, resulta fundamental para el trabajo porque permite obtener el valor óptimo en grafos pequeños. Ese valor óptimo se utiliza posteriormente como referencia para evaluar la calidad de la solución obtenida mediante búsqueda local.

###Solución mejorada: búsqueda local

Como solución mejorada se implementó una búsqueda local. A diferencia de la fuerza bruta, este enfoque no evalúa todas las particiones posibles, sino que parte desde una solución inicial y busca mejorarla mediante cambios pequeños.

En este caso, un movimiento local consiste en cambiar un vértice de un subconjunto al otro. Si este cambio aumenta el valor del corte, el movimiento se acepta. El proceso se repite hasta que no existan movimientos locales que mejoren la solución o hasta alcanzar el límite de iteraciones establecido.

La búsqueda local no garantiza encontrar siempre la solución óptima, ya que puede quedar atrapada en un óptimo local. Para reducir este riesgo, se utilizaron reinicios aleatorios. En cada reinicio se genera una nueva solución inicial y se aplica nuevamente el proceso de mejora local. Al final, se conserva la mejor solución encontrada entre todos los reinicios.

La complejidad aproximada de este enfoque puede expresarse como:

$$
O(R \cdot P \cdot n \cdot d)
$$

donde $R$ corresponde al número de reinicios, $P$ al número máximo de pasadas o iteraciones de mejora, $n$ al número de vértices y $d$ al grado promedio del grafo.

Este enfoque resulta más escalable que la fuerza bruta, ya que evita recorrer todo el espacio de soluciones. Su objetivo no es garantizar optimalidad absoluta, sino obtener soluciones de buena calidad en tiempos computacionalmente razonables.


### Generación de instancias

Las instancias utilizadas corresponden a grafos no dirigidos generados aleatoriamente. Cada grafo se define a partir de un número de vértices $n$ y una probabilidad $p$ de existencia de arista entre cada par de vértices.

Para cada par de vértices $(i,j)$, con $i < j$, se genera una arista con probabilidad $p$. De esta forma, valores bajos de $p$ producen grafos dispersos, mientras que valores altos producen grafos más densos.

En este trabajo se utilizaron grafos con distintos tamaños y tres niveles de densidad: disperso, medio y denso. Cada instancia fue guardada en un archivo `.txt` con el número de vértices, el número de aristas y el listado de aristas del grafo.

Este procedimiento permite evaluar el comportamiento de los algoritmos frente a distintos escenarios, observando cómo varía el tiempo de ejecución y la calidad de la solución cuando aumenta el tamaño o la densidad del grafo.
Se generan grafos aleatorios en formato `.txt`, porque este formato se puede leer de forma simple en Python, C++ y Java.

Cada archivo tiene la forma:

```text
n m
u v
u v
...
```

Además, se guarda un `metadata.csv` con información de cada instancia: cantidad de vértices, cantidad de aristas, densidad, semilla y ruta del archivo.


````python
import random
import pandas as pd
from pathlib import Path

INSTANCE_DIR = Path("maxcut_instances")
INSTANCE_DIR.mkdir(exist_ok=True)

METADATA_PATH = INSTANCE_DIR / "metadata.csv"

SIZES = [6, 8, 10, 12, 14, 16, 18, 20, 25, 30, 35, 40]
DENSITIES = [0.2, 0.5, 0.8]
REPETITIONS = 1
BASE_SEED = 42
BRUTEFORCE_MAX_N = 20


def generate_graph(n, p, seed):
    rng = random.Random(seed)
    edges = []

    for u in range(n):
        for v in range(u + 1, n):
            if rng.random() < p:
                edges.append((u, v))

    return edges


def save_graph_txt(file_path, n, edges):
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(f"{n} {len(edges)}\n")
        for u, v in edges:
            f.write(f"{u} {v}\n")


metadata = []

for n in SIZES:
    for p in DENSITIES:
        for rep in range(1, REPETITIONS + 1):
            seed = BASE_SEED + n * 1000 + int(p * 100) * 10 + rep
            edges = generate_graph(n, p, seed)

            while len(edges) == 0:
                seed += 1
                edges = generate_graph(n, p, seed)

            p_label = int(p * 100)
            instance_id = f"g_n{n:03d}_p{p_label}_r{rep}"
            file_path = INSTANCE_DIR / f"{instance_id}.txt"

            save_graph_txt(file_path, n, edges)

            metadata.append({
                "instance_id": instance_id,
                "file_path": str(file_path),
                "n": n,
                "m": len(edges),
                "p": p,
                "density": "disperso" if p == 0.2 else "medio" if p == 0.5 else "denso",
                "repetition": rep,
                "seed": seed,
                "use_bruteforce": n <= BRUTEFORCE_MAX_N,
                "use_local_search": True
            })

metadata_df = pd.DataFrame(metadata)
metadata_df.to_csv(METADATA_PATH, index=False)

print("Instancias generadas:", len(metadata_df))
metadata_df.head()
````

##Python

###Solucion base/ Fuerza bruta en python

Esta versión revisa todas las particiones posibles del grafo. Para evitar evaluar particiones equivalentes, se fija el vértice 0 en uno de los subconjuntos. Así se revisan $2^{n-1}$ particiones en lugar de $2^n$.

Esta solución se usa como referencia exacta para los grafos pequeños.

````python
import time
import pandas as pd
from pathlib import Path

INSTANCE_DIR = Path("maxcut_instances")
METADATA_PATH = INSTANCE_DIR / "metadata.csv"
RESULTS_PATH = "results_python_bruteforce.csv"

metadata_df = pd.read_csv(METADATA_PATH)


def load_graph_txt(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        n, m = map(int, f.readline().split())
        edges = []

        for line in f:
            if line.strip():
                u, v = map(int, line.split())
                edges.append((u, v))

    return {"n": n, "m": m, "edges": edges}


def vertex_side(mask, vertex):
    if vertex == 0:
        return 0
    return (mask >> (vertex - 1)) & 1


def cut_value_from_mask(edges, mask):
    cut = 0

    for u, v in edges:
        if vertex_side(mask, u) != vertex_side(mask, v):
            cut += 1

    return cut


def assignment_from_mask(n, mask):
    assignment = [0] * n

    for vertex in range(1, n):
        assignment[vertex] = (mask >> (vertex - 1)) & 1

    return assignment


def max_cut_bruteforce_python(graph, timeout_sec=60):
    n = graph["n"]
    edges = graph["edges"]

    start_time = time.perf_counter()

    best_cut = -1
    best_mask = 0

    iterations = 0
    total_partitions = 1 << (n - 1)
    check_interval = 1000

    for mask in range(total_partitions):
        iterations += 1

        current_cut = cut_value_from_mask(edges, mask)

        if current_cut > best_cut:
            best_cut = current_cut
            best_mask = mask

        if iterations % check_interval == 0:
            elapsed = time.perf_counter() - start_time

            if elapsed >= timeout_sec:
                return {
                    "cut_value": best_cut,
                    "assignment": assignment_from_mask(n, best_mask),
                    "status": "TIMEOUT",
                    "iterations": iterations,
                    "nodes_visited": iterations,
                    "total_partitions": total_partitions
                }

    return {
        "cut_value": best_cut,
        "assignment": assignment_from_mask(n, best_mask),
        "status": "OK",
        "iterations": iterations,
        "nodes_visited": iterations,
        "total_partitions": total_partitions
    }


def run_single_python_instance(instance_row, algorithm_func, algorithm_name, timeout_sec=60):
    graph = load_graph_txt(instance_row["file_path"])
    start_time = time.perf_counter()

    try:
        result = algorithm_func(graph, timeout_sec=timeout_sec)
        elapsed_time = time.perf_counter() - start_time

        return {
            "instance_id": instance_row["instance_id"],
            "algorithm": algorithm_name,
            "language": "Python",
            "n": int(instance_row["n"]),
            "m": int(instance_row["m"]),
            "p": float(instance_row["p"]),
            "density": instance_row["density"],
            "repetition": int(instance_row["repetition"]),
            "seed": int(instance_row["seed"]),
            "cut_value": result["cut_value"],
            "time_sec": elapsed_time,
            "status": result["status"],
            "iterations": result.get("iterations", "-"),
            "nodes_visited": result.get("nodes_visited", "-"),
            "total_partitions": result.get("total_partitions", "-"),
            "file_path": instance_row["file_path"]
        }

    except Exception as e:
        elapsed_time = time.perf_counter() - start_time

        return {
            "instance_id": instance_row["instance_id"],
            "algorithm": algorithm_name,
            "language": "Python",
            "n": int(instance_row["n"]),
            "m": int(instance_row["m"]),
            "p": float(instance_row["p"]),
            "density": instance_row["density"],
            "repetition": int(instance_row["repetition"]),
            "seed": int(instance_row["seed"]),
            "cut_value": "-",
            "time_sec": elapsed_time,
            "status": "ERROR",
            "iterations": "-",
            "nodes_visited": "-",
            "total_partitions": "-",
            "file_path": instance_row["file_path"],
            "error_message": str(e)
        }


def run_python_bruteforce_experiment(metadata_df, timeout_sec=60):
    selected_instances = metadata_df[metadata_df["use_bruteforce"] == True].copy()
    results = []

    for _, instance_row in selected_instances.iterrows():
        print(
            f"Ejecutando Python Fuerza Bruta | "
            f"{instance_row['instance_id']} | "
            f"n={instance_row['n']} | "
            f"m={instance_row['m']}"
        )

        results.append(
            run_single_python_instance(
                instance_row=instance_row,
                algorithm_func=max_cut_bruteforce_python,
                algorithm_name="Bruteforce",
                timeout_sec=timeout_sec
            )
        )

    results_df = pd.DataFrame(results)
    results_df.to_csv(RESULTS_PATH, index=False)

    return results_df


results_python_bruteforce_df = run_python_bruteforce_experiment(metadata_df, timeout_sec=60)
results_python_bruteforce_df.head()
````

##Búsqueda local en Python

La búsqueda local se usa como solución mejorada. A diferencia de fuerza bruta, no revisa todas las particiones, sino que parte desde una solución inicial y mueve vértices cuando eso mejora el corte.

Como puede quedar atrapada en óptimos locales, se usan varios reinicios y se conserva el mejor resultado.

````python
import time
import pandas as pd
from pathlib import Path

INSTANCE_DIR = Path("maxcut_instances")
METADATA_PATH = INSTANCE_DIR / "metadata.csv"
RESULTS_LOCAL_SEARCH_PYTHON_PATH = "results_python_local_search.csv"

metadata_df = pd.read_csv(METADATA_PATH)


def build_adjacency_list(n, edges):
    adjacency = [[] for _ in range(n)]

    for u, v in edges:
        adjacency[u].append(v)
        adjacency[v].append(u)

    return adjacency


def calculate_cut_value(edges, assignment):
    cut = 0

    for u, v in edges:
        if assignment[u] != assignment[v]:
            cut += 1

    return cut


def lcg_next(state):
    return (1103515245 * state + 12345) & 0x7fffffff


def generate_initial_assignment(n, seed):
    assignment = [0] * n
    state = seed & 0x7fffffff

    for vertex in range(1, n):
        state = lcg_next(state)
        assignment[vertex] = state % 2

    return assignment


def flip_gain(vertex, assignment, adjacency):
    same_side = 0
    opposite_side = 0

    for neighbor in adjacency[vertex]:
        if assignment[vertex] == assignment[neighbor]:
            same_side += 1
        else:
            opposite_side += 1

    return same_side - opposite_side


def max_cut_local_search_python(
    graph,
    timeout_sec=60,
    restarts=20,
    max_passes=1000,
    base_seed=42
):
    n = graph["n"]
    edges = graph["edges"]
    adjacency = build_adjacency_list(n, edges)

    start_time = time.perf_counter()

    best_cut = -1
    best_assignment = None

    total_iterations = 0
    total_nodes_visited = 0
    completed_restarts = 0

    for restart in range(restarts):
        elapsed = time.perf_counter() - start_time

        if elapsed >= timeout_sec:
            return {
                "cut_value": best_cut,
                "assignment": best_assignment,
                "status": "TIMEOUT",
                "iterations": total_iterations,
                "nodes_visited": total_nodes_visited,
                "restarts": completed_restarts
            }

        assignment = generate_initial_assignment(n, base_seed + restart * 1009)
        current_cut = calculate_cut_value(edges, assignment)

        improved = True
        passes = 0

        while improved and passes < max_passes:
            improved = False
            passes += 1
            total_iterations += 1

            best_gain = 0
            best_vertex = -1

            for vertex in range(1, n):
                total_nodes_visited += 1
                gain = flip_gain(vertex, assignment, adjacency)

                if gain > best_gain:
                    best_gain = gain
                    best_vertex = vertex

            if best_vertex != -1:
                assignment[best_vertex] = 1 - assignment[best_vertex]
                current_cut += best_gain
                improved = True

            if total_iterations % 100 == 0:
                elapsed = time.perf_counter() - start_time

                if elapsed >= timeout_sec:
                    return {
                        "cut_value": best_cut,
                        "assignment": best_assignment,
                        "status": "TIMEOUT",
                        "iterations": total_iterations,
                        "nodes_visited": total_nodes_visited,
                        "restarts": completed_restarts
                    }

        completed_restarts += 1

        if current_cut > best_cut:
            best_cut = current_cut
            best_assignment = assignment.copy()

    return {
        "cut_value": best_cut,
        "assignment": best_assignment,
        "status": "OK",
        "iterations": total_iterations,
        "nodes_visited": total_nodes_visited,
        "restarts": completed_restarts
    }


def run_single_python_local_search_instance(instance_row, timeout_sec=60, restarts=20, max_passes=1000):
    graph = load_graph_txt(instance_row["file_path"])
    start_time = time.perf_counter()

    try:
        result = max_cut_local_search_python(
            graph=graph,
            timeout_sec=timeout_sec,
            restarts=restarts,
            max_passes=max_passes,
            base_seed=int(instance_row["seed"])
        )

        elapsed_time = time.perf_counter() - start_time

        return {
            "instance_id": instance_row["instance_id"],
            "algorithm": "Local Search",
            "language": "Python",
            "n": int(instance_row["n"]),
            "m": int(instance_row["m"]),
            "p": float(instance_row["p"]),
            "density": instance_row["density"],
            "repetition": int(instance_row["repetition"]),
            "seed": int(instance_row["seed"]),
            "cut_value": result["cut_value"],
            "time_sec": elapsed_time,
            "status": result["status"],
            "iterations": result.get("iterations", "-"),
            "nodes_visited": result.get("nodes_visited", "-"),
            "restarts": result.get("restarts", "-"),
            "file_path": instance_row["file_path"]
        }

    except Exception as e:
        elapsed_time = time.perf_counter() - start_time

        return {
            "instance_id": instance_row["instance_id"],
            "algorithm": "Local Search",
            "language": "Python",
            "n": int(instance_row["n"]),
            "m": int(instance_row["m"]),
            "p": float(instance_row["p"]),
            "density": instance_row["density"],
            "repetition": int(instance_row["repetition"]),
            "seed": int(instance_row["seed"]),
            "cut_value": "-",
            "time_sec": elapsed_time,
            "status": "ERROR",
            "iterations": "-",
            "nodes_visited": "-",
            "restarts": "-",
            "file_path": instance_row["file_path"],
            "error_message": str(e)
        }


def run_python_local_search_experiment(metadata_df, timeout_sec=60, restarts=20, max_passes=1000):
    results = []

    for _, instance_row in metadata_df.iterrows():
        print(
            f"Ejecutando Python Búsqueda Local | "
            f"{instance_row['instance_id']} | "
            f"n={instance_row['n']} | "
            f"m={instance_row['m']}"
        )

        results.append(
            run_single_python_local_search_instance(
                instance_row=instance_row,
                timeout_sec=timeout_sec,
                restarts=restarts,
                max_passes=max_passes
            )
        )

    results_df = pd.DataFrame(results)
    results_df.to_csv(RESULTS_LOCAL_SEARCH_PYTHON_PATH, index=False)

    return results_df


results_python_local_search_df = run_python_local_search_experiment(
    metadata_df,
    timeout_sec=60,
    restarts=20,
    max_passes=1000
)


results_python_local_search_df.head()
````

###Fuerza bruta en C++

La implementación en C++ usa el mismo formato de entrada y entrega una línea CSV con el resultado. Luego Python ejecuta el programa, captura la salida y la guarda en un archivo de resultados.

````python
%%writefile maxcut_bruteforce.cpp
#include <iostream>
#include <fstream>
#include <vector>
#include <string>
#include <chrono>
#include <iomanip>
#include <stdexcept>

using namespace std;

struct Graph {
    int n;
    int m;
    vector<pair<int, int>> edges;
};

struct Result {
    long long cut_value;
    double time_sec;
    string status;
    unsigned long long iterations;
    unsigned long long nodes_visited;
    unsigned long long total_partitions;
};

Graph load_graph_txt(const string& file_path) {
    ifstream file(file_path);

    if (!file.is_open()) {
        throw runtime_error("No se pudo abrir el archivo: " + file_path);
    }

    Graph graph;
    file >> graph.n >> graph.m;

    for (int i = 0; i < graph.m; i++) {
        int u, v;
        file >> u >> v;
        graph.edges.push_back({u, v});
    }

    return graph;
}

int vertex_side(unsigned long long mask, int vertex) {
    if (vertex == 0) {
        return 0;
    }

    return (mask >> (vertex - 1)) & 1ULL;
}

long long cut_value_from_mask(const vector<pair<int, int>>& edges, unsigned long long mask) {
    long long cut = 0;

    for (const auto& edge : edges) {
        if (vertex_side(mask, edge.first) != vertex_side(mask, edge.second)) {
            cut++;
        }
    }

    return cut;
}

Result max_cut_bruteforce_cpp(const Graph& graph, double timeout_sec) {
    int n = graph.n;

    if (n <= 0) {
        throw runtime_error("El grafo debe tener al menos un vertice.");
    }

    if (n > 63) {
        throw runtime_error("n demasiado grande para usar mascara de 64 bits.");
    }

    auto start_time = chrono::high_resolution_clock::now();

    long long best_cut = -1;
    unsigned long long iterations = 0;
    unsigned long long total_partitions = 1ULL << (n - 1);

    const unsigned long long check_interval = 1000;

    for (unsigned long long mask = 0; mask < total_partitions; mask++) {
        iterations++;

        long long current_cut = cut_value_from_mask(graph.edges, mask);

        if (current_cut > best_cut) {
            best_cut = current_cut;
        }

        if (iterations % check_interval == 0) {
            auto current_time = chrono::high_resolution_clock::now();
            chrono::duration<double> elapsed = current_time - start_time;

            if (elapsed.count() >= timeout_sec) {
                return {best_cut, elapsed.count(), "TIMEOUT", iterations, iterations, total_partitions};
            }
        }
    }

    auto end_time = chrono::high_resolution_clock::now();
    chrono::duration<double> elapsed = end_time - start_time;

    return {best_cut, elapsed.count(), "OK", iterations, iterations, total_partitions};
}

int main(int argc, char* argv[]) {
    if (argc < 2) {
        cerr << "Uso: ./maxcut_bruteforce_cpp archivo.txt [timeout_segundos]" << endl;
        return 1;
    }

    string file_path = argv[1];
    double timeout_sec = (argc >= 3) ? stod(argv[2]) : 60.0;

    try {
        Graph graph = load_graph_txt(file_path);
        Result result = max_cut_bruteforce_cpp(graph, timeout_sec);

        cout << result.cut_value << ","
             << fixed << setprecision(6) << result.time_sec << ","
             << result.status << ","
             << result.iterations << ","
             << result.nodes_visited << ","
             << result.total_partitions
             << endl;

    } catch (const exception& e) {
        cerr << "ERROR: " << e.what() << endl;
        return 1;
    }

    return 0;
}
````

Esta versión replica la lógica usada en Python. Se usa el mismo generador inicial, los mismos parámetros principales y la misma estructura de salida para poder comparar los resultados.

````python
%%writefile maxcut_local_search.cpp
#include <iostream>
#include <fstream>
#include <vector>
#include <string>
#include <chrono>
#include <iomanip>
#include <stdexcept>

using namespace std;

struct Graph {
    int n;
    int m;
    vector<pair<int, int>> edges;
};

struct Result {
    long long cut_value;
    double time_sec;
    string status;
    long long iterations;
    long long nodes_visited;
    int restarts;
};

Graph load_graph_txt(const string& file_path) {
    ifstream file(file_path);
    if (!file.is_open()) {
        throw runtime_error("No se pudo abrir el archivo: " + file_path);
    }

    Graph graph;
    file >> graph.n >> graph.m;

    for (int i = 0; i < graph.m; i++) {
        int u, v;
        file >> u >> v;
        graph.edges.push_back({u, v});
    }

    return graph;
}

vector<vector<int>> build_adjacency_list(int n, const vector<pair<int, int>>& edges) {
    vector<vector<int>> adjacency(n);

    for (const auto& edge : edges) {
        adjacency[edge.first].push_back(edge.second);
        adjacency[edge.second].push_back(edge.first);
    }

    return adjacency;
}

long long calculate_cut_value(const vector<pair<int, int>>& edges, const vector<int>& assignment) {
    long long cut = 0;

    for (const auto& edge : edges) {
        if (assignment[edge.first] != assignment[edge.second]) {
            cut++;
        }
    }

    return cut;
}

long long lcg_next(long long state) {
    return (1103515245LL * state + 12345LL) & 0x7fffffffLL;
}

vector<int> generate_initial_assignment(int n, long long seed) {
    vector<int> assignment(n, 0);
    long long state = seed & 0x7fffffffLL;

    for (int vertex = 1; vertex < n; vertex++) {
        state = lcg_next(state);
        assignment[vertex] = state % 2;
    }

    return assignment;
}

int flip_gain(int vertex, const vector<int>& assignment, const vector<vector<int>>& adjacency) {
    int same_side = 0;
    int opposite_side = 0;

    for (int neighbor : adjacency[vertex]) {
        if (assignment[vertex] == assignment[neighbor]) {
            same_side++;
        } else {
            opposite_side++;
        }
    }

    return same_side - opposite_side;
}

Result max_cut_local_search_cpp(
    const Graph& graph,
    double timeout_sec,
    int restarts,
    int max_passes,
    long long base_seed
) {
    int n = graph.n;
    vector<vector<int>> adjacency = build_adjacency_list(n, graph.edges);

    auto start_time = chrono::high_resolution_clock::now();

    long long best_cut = -1;
    long long total_iterations = 0;
    long long total_nodes_visited = 0;
    int completed_restarts = 0;

    for (int restart = 0; restart < restarts; restart++) {
        auto current_time = chrono::high_resolution_clock::now();
        chrono::duration<double> elapsed = current_time - start_time;

        if (elapsed.count() >= timeout_sec) {
            return {best_cut, elapsed.count(), "TIMEOUT", total_iterations, total_nodes_visited, completed_restarts};
        }

        vector<int> assignment = generate_initial_assignment(n, base_seed + restart * 1009LL);
        long long current_cut = calculate_cut_value(graph.edges, assignment);

        bool improved = true;
        int passes = 0;

        while (improved && passes < max_passes) {
            improved = false;
            passes++;
            total_iterations++;

            int best_gain = 0;
            int best_vertex = -1;

            for (int vertex = 1; vertex < n; vertex++) {
                total_nodes_visited++;
                int gain = flip_gain(vertex, assignment, adjacency);

                if (gain > best_gain) {
                    best_gain = gain;
                    best_vertex = vertex;
                }
            }

            if (best_vertex != -1) {
                assignment[best_vertex] = 1 - assignment[best_vertex];
                current_cut += best_gain;
                improved = true;
            }

            if (total_iterations % 100 == 0) {
                current_time = chrono::high_resolution_clock::now();
                elapsed = current_time - start_time;

                if (elapsed.count() >= timeout_sec) {
                    return {best_cut, elapsed.count(), "TIMEOUT", total_iterations, total_nodes_visited, completed_restarts};
                }
            }
        }

        completed_restarts++;

        if (current_cut > best_cut) {
            best_cut = current_cut;
        }
    }

    auto end_time = chrono::high_resolution_clock::now();
    chrono::duration<double> elapsed = end_time - start_time;

    return {best_cut, elapsed.count(), "OK", total_iterations, total_nodes_visited, completed_restarts};
}

int main(int argc, char* argv[]) {
    if (argc < 2) {
        cerr << "Uso: ./maxcut_local_search_cpp archivo.txt [timeout] [restarts] [max_passes] [seed]" << endl;
        return 1;
    }

    string file_path = argv[1];

    double timeout_sec = (argc >= 3) ? stod(argv[2]) : 60.0;
    int restarts = (argc >= 4) ? stoi(argv[3]) : 20;
    int max_passes = (argc >= 5) ? stoi(argv[4]) : 1000;
    long long seed = (argc >= 6) ? stoll(argv[5]) : 42;

    try {
        Graph graph = load_graph_txt(file_path);
        Result result = max_cut_local_search_cpp(graph, timeout_sec, restarts, max_passes, seed);

        cout << result.cut_value << ","
             << fixed << setprecision(6) << result.time_sec << ","
             << result.status << ","
             << result.iterations << ","
             << result.nodes_visited << ","
             << result.restarts
             << endl;

    } catch (const exception& e) {
        cerr << "ERROR: " << e.what() << endl;
        return 1;
    }

    return 0;
}
````

## Fuerza bruta en Java

Java sigue la misma estructura que C++: lee el archivo `.txt`, ejecuta fuerza bruta, mide el tiempo y entrega una línea CSV con el resultado.

````python
%%writefile MaxCutBruteforce.java
import java.io.*;
import java.util.*;
import java.util.Locale;

public class MaxCutBruteforce {
    static class Graph {
        int n;
        int m;
        int[] u;
        int[] v;

        Graph(int n, int m, int[] u, int[] v) {
            this.n = n;
            this.m = m;
            this.u = u;
            this.v = v;
        }
    }

    static class Result {
        long cutValue;
        double timeSec;
        String status;
        long iterations;
        long nodesVisited;
        long totalPartitions;

        Result(long cutValue, double timeSec, String status, long iterations, long nodesVisited, long totalPartitions) {
            this.cutValue = cutValue;
            this.timeSec = timeSec;
            this.status = status;
            this.iterations = iterations;
            this.nodesVisited = nodesVisited;
            this.totalPartitions = totalPartitions;
        }
    }

    static Graph loadGraphTxt(String filePath) throws IOException {
        BufferedReader br = new BufferedReader(new FileReader(filePath));
        String firstLine = br.readLine();

        if (firstLine == null || firstLine.trim().isEmpty()) {
            br.close();
            throw new IOException("Archivo vacio o invalido.");
        }

        StringTokenizer st = new StringTokenizer(firstLine);
        int n = Integer.parseInt(st.nextToken());
        int m = Integer.parseInt(st.nextToken());

        int[] u = new int[m];
        int[] v = new int[m];

        for (int i = 0; i < m; i++) {
            String line = br.readLine();

            if (line == null) {
                br.close();
                throw new IOException("El archivo tiene menos aristas que las declaradas.");
            }

            st = new StringTokenizer(line);
            u[i] = Integer.parseInt(st.nextToken());
            v[i] = Integer.parseInt(st.nextToken());
        }

        br.close();
        return new Graph(n, m, u, v);
    }

    static int vertexSide(long mask, int vertex) {
        if (vertex == 0) {
            return 0;
        }

        return (int) ((mask >> (vertex - 1)) & 1L);
    }

    static long cutValueFromMask(Graph graph, long mask) {
        long cut = 0;

        for (int i = 0; i < graph.m; i++) {
            if (vertexSide(mask, graph.u[i]) != vertexSide(mask, graph.v[i])) {
                cut++;
            }
        }

        return cut;
    }

    static Result maxCutBruteforce(Graph graph, double timeoutSec) {
        int n = graph.n;

        if (n <= 0) {
            throw new IllegalArgumentException("El grafo debe tener al menos un vertice.");
        }

        if (n > 63) {
            throw new IllegalArgumentException("n demasiado grande para usar mascaras de 64 bits.");
        }

        long startTime = System.nanoTime();

        long bestCut = -1;
        long iterations = 0;
        long totalPartitions = 1L << (n - 1);

        final long checkInterval = 1000;

        for (long mask = 0; mask < totalPartitions; mask++) {
            iterations++;

            long currentCut = cutValueFromMask(graph, mask);

            if (currentCut > bestCut) {
                bestCut = currentCut;
            }

            if (iterations % checkInterval == 0) {
                double elapsedSec = (System.nanoTime() - startTime) / 1_000_000_000.0;

                if (elapsedSec >= timeoutSec) {
                    return new Result(bestCut, elapsedSec, "TIMEOUT", iterations, iterations, totalPartitions);
                }
            }
        }

        double elapsedSec = (System.nanoTime() - startTime) / 1_000_000_000.0;

        return new Result(bestCut, elapsedSec, "OK", iterations, iterations, totalPartitions);
    }

    public static void main(String[] args) {
        Locale.setDefault(Locale.US);

        if (args.length < 1) {
            System.err.println("Uso: java MaxCutBruteforce archivo.txt [timeout_segundos]");
            System.exit(1);
        }

        String filePath = args[0];
        double timeoutSec = (args.length >= 2) ? Double.parseDouble(args[1]) : 60.0;

        try {
            Graph graph = loadGraphTxt(filePath);
            Result result = maxCutBruteforce(graph, timeoutSec);

            System.out.printf(
                Locale.US,
                "%d,%.6f,%s,%d,%d,%d%n",
                result.cutValue,
                result.timeSec,
                result.status,
                result.iterations,
                result.nodesVisited,
                result.totalPartitions
            );

        } catch (Exception e) {
            System.err.println("ERROR: " + e.getMessage());
            System.exit(1);
        }
    }
}
````

###Java — Búsqueda local

Esta implementación completa la solución mejorada en los tres lenguajes. Java lee las mismas instancias `.txt`, aplica búsqueda local con reinicios y devuelve una línea CSV con el resultado de cada ejecución.

````python
%%writefile MaxCutLocalSearch.java
import java.io.*;
import java.util.*;
import java.util.Locale;

public class MaxCutLocalSearch {

    static class Graph {
        int n;
        int m;
        int[] u;
        int[] v;

        Graph(int n, int m, int[] u, int[] v) {
            this.n = n;
            this.m = m;
            this.u = u;
            this.v = v;
        }
    }

    static class Result {
        long cutValue;
        double timeSec;
        String status;
        long iterations;
        long nodesVisited;
        int restarts;

        Result(long cutValue, double timeSec, String status, long iterations, long nodesVisited, int restarts) {
            this.cutValue = cutValue;
            this.timeSec = timeSec;
            this.status = status;
            this.iterations = iterations;
            this.nodesVisited = nodesVisited;
            this.restarts = restarts;
        }
    }

    static Graph loadGraphTxt(String filePath) throws IOException {
        BufferedReader br = new BufferedReader(new FileReader(filePath));
        String firstLine = br.readLine();

        if (firstLine == null || firstLine.trim().isEmpty()) {
            br.close();
            throw new IOException("Archivo vacio o invalido.");
        }

        StringTokenizer st = new StringTokenizer(firstLine);
        int n = Integer.parseInt(st.nextToken());
        int m = Integer.parseInt(st.nextToken());

        int[] u = new int[m];
        int[] v = new int[m];

        for (int i = 0; i < m; i++) {
            String line = br.readLine();

            if (line == null) {
                br.close();
                throw new IOException("El archivo tiene menos aristas que las declaradas.");
            }

            st = new StringTokenizer(line);
            u[i] = Integer.parseInt(st.nextToken());
            v[i] = Integer.parseInt(st.nextToken());
        }

        br.close();
        return new Graph(n, m, u, v);
    }

    @SuppressWarnings("unchecked")
    static ArrayList<Integer>[] buildAdjacencyList(Graph graph) {
        ArrayList<Integer>[] adjacency = new ArrayList[graph.n];

        for (int i = 0; i < graph.n; i++) {
            adjacency[i] = new ArrayList<>();
        }

        for (int i = 0; i < graph.m; i++) {
            int a = graph.u[i];
            int b = graph.v[i];

            adjacency[a].add(b);
            adjacency[b].add(a);
        }

        return adjacency;
    }

    static long calculateCutValue(Graph graph, int[] assignment) {
        long cut = 0;

        for (int i = 0; i < graph.m; i++) {
            if (assignment[graph.u[i]] != assignment[graph.v[i]]) {
                cut++;
            }
        }

        return cut;
    }

    static long lcgNext(long state) {
        return (1103515245L * state + 12345L) & 0x7fffffffL;
    }

    static int[] generateInitialAssignment(int n, long seed) {
        int[] assignment = new int[n];
        long state = seed & 0x7fffffffL;

        assignment[0] = 0;

        for (int vertex = 1; vertex < n; vertex++) {
            state = lcgNext(state);
            assignment[vertex] = (int) (state % 2);
        }

        return assignment;
    }

    static int flipGain(int vertex, int[] assignment, ArrayList<Integer>[] adjacency) {
        int sameSide = 0;
        int oppositeSide = 0;

        for (int neighbor : adjacency[vertex]) {
            if (assignment[vertex] == assignment[neighbor]) {
                sameSide++;
            } else {
                oppositeSide++;
            }
        }

        return sameSide - oppositeSide;
    }

    static Result maxCutLocalSearch(Graph graph, double timeoutSec, int restarts, int maxPasses, long baseSeed) {
        int n = graph.n;

        if (n <= 0) {
            throw new IllegalArgumentException("El grafo debe tener al menos un vertice.");
        }

        ArrayList<Integer>[] adjacency = buildAdjacencyList(graph);
        long startTime = System.nanoTime();

        long bestCut = -1;
        long totalIterations = 0;
        long totalNodesVisited = 0;
        int completedRestarts = 0;

        for (int restart = 0; restart < restarts; restart++) {
            double elapsedSec = (System.nanoTime() - startTime) / 1_000_000_000.0;

            if (elapsedSec >= timeoutSec) {
                return new Result(bestCut, elapsedSec, "TIMEOUT", totalIterations, totalNodesVisited, completedRestarts);
            }

            long currentSeed = baseSeed + restart * 1009L;
            int[] assignment = generateInitialAssignment(n, currentSeed);
            long currentCut = calculateCutValue(graph, assignment);

            boolean improved = true;
            int passes = 0;

            while (improved && passes < maxPasses) {
                improved = false;
                passes++;
                totalIterations++;

                int bestGain = 0;
                int bestVertex = -1;

                for (int vertex = 1; vertex < n; vertex++) {
                    totalNodesVisited++;
                    int gain = flipGain(vertex, assignment, adjacency);

                    if (gain > bestGain) {
                        bestGain = gain;
                        bestVertex = vertex;
                    }
                }

                if (bestVertex != -1) {
                    assignment[bestVertex] = 1 - assignment[bestVertex];
                    currentCut += bestGain;
                    improved = true;
                }

                if (totalIterations % 100 == 0) {
                    elapsedSec = (System.nanoTime() - startTime) / 1_000_000_000.0;

                    if (elapsedSec >= timeoutSec) {
                        return new Result(bestCut, elapsedSec, "TIMEOUT", totalIterations, totalNodesVisited, completedRestarts);
                    }
                }
            }

            completedRestarts++;

            if (currentCut > bestCut) {
                bestCut = currentCut;
            }
        }

        double elapsedSec = (System.nanoTime() - startTime) / 1_000_000_000.0;
        return new Result(bestCut, elapsedSec, "OK", totalIterations, totalNodesVisited, completedRestarts);
    }

    public static void main(String[] args) {
        Locale.setDefault(Locale.US);

        if (args.length < 1) {
            System.err.println("Uso: java MaxCutLocalSearch archivo.txt [timeout] [restarts] [max_passes] [seed]");
            System.exit(1);
        }

        String filePath = args[0];
        double timeoutSec = 60.0;
        int restarts = 20;
        int maxPasses = 1000;
        long seed = 42;

        if (args.length >= 2) timeoutSec = Double.parseDouble(args[1]);
        if (args.length >= 3) restarts = Integer.parseInt(args[2]);
        if (args.length >= 4) maxPasses = Integer.parseInt(args[3]);
        if (args.length >= 5) seed = Long.parseLong(args[4]);

        try {
            Graph graph = loadGraphTxt(filePath);
            Result result = maxCutLocalSearch(graph, timeoutSec, restarts, maxPasses, seed);

            System.out.printf(
                Locale.US,
                "%d,%.6f,%s,%d,%d,%d%n",
                result.cutValue,
                result.timeSec,
                result.status,
                result.iterations,
                result.nodesVisited,
                result.restarts
            );

        } catch (Exception e) {
            System.err.println("ERROR: " + e.getMessage());
            System.exit(1);
        }
    }
}
````

````python
import subprocess
import pandas as pd
from pathlib import Path

INSTANCE_DIR = Path("maxcut_instances")
METADATA_PATH = INSTANCE_DIR / "metadata.csv"

JAVA_LOCAL_SEARCH_CLASS = "MaxCutLocalSearch"
RESULTS_JAVA_LOCAL_SEARCH_PATH = "results_java_local_search.csv"

metadata_df = pd.read_csv(METADATA_PATH)


def parse_java_local_search_output(output):
    parts = output.strip().split(",")

    if len(parts) != 6:
        raise ValueError(f"Salida inesperada desde Java: {output}")

    return {
        "cut_value": int(parts[0]),
        "time_sec": float(parts[1]),
        "status": parts[2],
        "iterations": int(parts[3]),
        "nodes_visited": int(parts[4]),
        "restarts": int(parts[5])
    }


def run_single_java_local_search_instance(instance_row, timeout_sec=60, restarts=20, max_passes=1000):
    command = [
        "java",
        JAVA_LOCAL_SEARCH_CLASS,
        instance_row["file_path"],
        str(timeout_sec),
        str(restarts),
        str(max_passes),
        str(int(instance_row["seed"]))
    ]

    try:
        completed_process = subprocess.run(
            command,
            capture_output=True,
            text=True,
            timeout=timeout_sec + 5
        )

        base_row = {
            "instance_id": instance_row["instance_id"],
            "algorithm": "Local Search",
            "language": "Java",
            "n": int(instance_row["n"]),
            "m": int(instance_row["m"]),
            "p": float(instance_row["p"]),
            "density": instance_row["density"],
            "repetition": int(instance_row["repetition"]),
            "seed": int(instance_row["seed"]),
            "file_path": instance_row["file_path"]
        }

        if completed_process.returncode != 0:
            return {
                **base_row,
                "cut_value": "-",
                "time_sec": "-",
                "status": "ERROR",
                "iterations": "-",
                "nodes_visited": "-",
                "restarts": "-",
                "error_message": completed_process.stderr.strip()
            }

        parsed = parse_java_local_search_output(completed_process.stdout)

        return {
            **base_row,
            "cut_value": parsed["cut_value"],
            "time_sec": parsed["time_sec"],
            "status": parsed["status"],
            "iterations": parsed["iterations"],
            "nodes_visited": parsed["nodes_visited"],
            "restarts": parsed["restarts"]
        }

    except subprocess.TimeoutExpired:
        return {
            "instance_id": instance_row["instance_id"],
            "algorithm": "Local Search",
            "language": "Java",
            "n": int(instance_row["n"]),
            "m": int(instance_row["m"]),
            "p": float(instance_row["p"]),
            "density": instance_row["density"],
            "repetition": int(instance_row["repetition"]),
            "seed": int(instance_row["seed"]),
            "cut_value": "-",
            "time_sec": timeout_sec,
            "status": "PYTHON_TIMEOUT",
            "iterations": "-",
            "nodes_visited": "-",
            "restarts": "-",
            "file_path": instance_row["file_path"],
            "error_message": "Timeout externo desde Python"
        }


def run_java_local_search_experiment(metadata_df, timeout_sec=60, restarts=20, max_passes=1000):
    selected_instances = metadata_df.copy()
    results = []

    for _, instance_row in selected_instances.iterrows():
        print(
            f"Ejecutando Java Búsqueda Local | "
            f"{instance_row['instance_id']} | "
            f"n={instance_row['n']} | "
            f"m={instance_row['m']} | "
            f"p={instance_row['p']}"
        )

        results.append(
            run_single_java_local_search_instance(
                instance_row=instance_row,
                timeout_sec=timeout_sec,
                restarts=restarts,
                max_passes=max_passes
            )
        )

    results_df = pd.DataFrame(results)
    results_df.to_csv(RESULTS_JAVA_LOCAL_SEARCH_PATH, index=False)

    print(f"Resultados Java búsqueda local guardados en: {RESULTS_JAVA_LOCAL_SEARCH_PATH}")
    return results_df


results_java_local_search_df = run_java_local_search_experiment(
    metadata_df,
    timeout_sec=60,
    restarts=20,
    max_passes=1000
)

results_java_local_search_df.head()
````

##Limpieza y columnas derivadas

Se realizó un proceso de limpieza y estandarización de los resultados experimentales con el objetivo de unificar los datos provenientes de distintos algoritmos y lenguajes.
````python
df = results_all_df.copy()

df["algorithm"] = df["algorithm"].replace({
    "Bruteforce": "Fuerza bruta",
    "Local Search": "Búsqueda local"
})

df["status"] = df["status"].astype(str).str.upper()

numeric_columns = [
    "n", "m", "p", "cut_value", "time_sec", "iterations",
    "nodes_visited", "restarts", "total_partitions"
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
    else:
        df[col] = np.nan

df["solved"] = df["status"] == "OK"
df["timeout"] = df["status"].str.contains("TIMEOUT", na=False)
df["error"] = df["status"].str.contains("ERROR", na=False)
df["cut_ratio"] = df["cut_value"] / df["m"]

language_order = ["Python", "C++", "Java"]
algorithm_order = ["Fuerza bruta", "Búsqueda local"]

df["language"] = pd.Categorical(df["language"], categories=language_order, ordered=True)
df["algorithm"] = pd.Categorical(df["algorithm"], categories=algorithm_order, ordered=True)

df = df.sort_values(["algorithm", "language", "n", "p", "instance_id"])
df.to_csv("results_all_clean.csv", index=False)

df.head()
````

## Tiempo de ejecucion

Los resultados experimentales evidencian diferencias significativas en los tiempos de ejecución entre los algoritmos evaluados. El algoritmo de fuerza bruta presenta tiempos considerablemente mayores, especialmente en implementaciones en Python, donde el costo computacional se vuelve elevado incluso para tamaños moderados de grafo. En contraste, la búsqueda local logra resolver las mismas instancias en tiempos considerablemente menores, manteniéndose en el orden de milisegundos en la mayoría de los casos.

Esta diferencia se explica por la naturaleza de ambos enfoques: mientras la fuerza bruta evalúa exhaustivamente todas las particiones posibles del grafo, la búsqueda local explora únicamente una fracción del espacio de soluciones mediante mejoras iterativas. Como resultado, la búsqueda local presenta ventajas claras en términos de eficiencia computacional, alcanzando mejoras de varios órdenes de magnitud en comparación con la fuerza bruta.

##Escalabilidad

El análisis de escalabilidad muestra que el algoritmo de fuerza bruta presenta un crecimiento exponencial en el tiempo de ejecución a medida que aumenta el número de vértices del grafo. ## Diferencias entre enfoques

La comparación directa entre fuerza bruta y búsqueda local pone en evidencia un trade-off fundamental entre optimalidad y eficiencia. La fuerza bruta garantiza encontrar la solución óptima del problema, ya que explora exhaustivamente el espacio de soluciones. Sin embargo, esta garantía tiene un alto costo computacional.

En contraste, la búsqueda local no asegura alcanzar la solución óptima, ya que puede quedar atrapada en óptimos locales. No obstante, los resultados experimentales muestran que, en la mayoría de los casos evaluados, la calidad de las soluciones obtenidas por la búsqueda local es cercana al óptimo, con diferencias relativamente pequeñas.

La diferencia significativa en el tiempo de ejecución entre ambos enfoques puede explicarse a partir de sus complejidades algorítmicas. El algoritmo de fuerza bruta tiene una complejidad del orden $O(2^{n−1} \cdot m)$, ya que evalúa todas las posibles particiones del conjunto de vértices, y para cada una calcula el valor del corte recorriendo las aristas del grafo. En contraste, la búsqueda local presenta una complejidad aproximada de $O(R \cdot P \cdot n \cdot d)$, donde $R$ corresponde al número de reinicios, $P$ al número de iteraciones por búsqueda, $n$ al número de vértices y $d$ al grado promedio del grafo. Dado que estos factores crecen de manera polinomial o se mantienen acotados por parámetros del algoritmo, la búsqueda local evita la explosión combinatoria característica de la fuerza bruta. Esta diferencia estructural explica por qué, a medida que aumenta el tamaño del grafo, la búsqueda local puede ser cientos o miles de veces más rápida, manteniendo al mismo tiempo una calidad de solución cercana al óptimo.

## Diferencia entre lenguajes

El rendimiento de los algoritmos también varía según el lenguaje de programación utilizado. En general, C++ presenta los mejores tiempos de ejecución, seguido por Java, mientras que Python resulta ser el más lento. Estas diferencias se deben a características propias de cada lenguaje, como el nivel de abstracción, el tipo de compilación y la eficiencia en la ejecución de operaciones básicas.

Sin embargo, es importante destacar que estas diferencias no afectan la complejidad algorítmica de los enfoques, sino que corresponden a factores de implementación. En todos los casos, el comportamiento general de los algoritmos se mantiene consistente, reflejando que la complejidad esencial del problema es independiente del lenguaje utilizado.

## Análisis critico

Si bien los resultados obtenidos son consistentes y permiten extraer conclusiones claras, es importante considerar ciertas limitaciones del estudio. En primer lugar, se utilizó una única repetición por instancia, lo que limita el análisis de variabilidad en los resultados, especialmente en el caso de la búsqueda local, que depende de condiciones iniciales aleatorias.

Además, los grafos generados no necesariamente representan los casos más adversos del problema, por lo que el desempeño observado de la búsqueda local podría variar en escenarios más complejos o estructurados. A pesar de ello, los resultados obtenidos son representativos del comportamiento general de los algoritmos en instancias típicas.

Finalmente, este análisis experimental confirma en la práctica la naturaleza NP-completa del problema Max-Cut, evidenciada en el crecimiento exponencial del algoritmo de fuerza bruta. Asimismo, resalta la importancia de utilizar heurísticas como la búsqueda local para obtener soluciones de buena calidad en tiempos computacionalmente razonables.
